```python
def calculator():
    x=int(input('Enter of the  number: '))
    y=int(input('Enter of the number: '))
    operators=input('Entre of the opretors: ')
    if operators=='+':
        print(f'The sum of the numbers:{x+y}')
    elif  operators=='-':
        print(f'The result of subtracting the numbers:{x-y}')
    elif  operators=='*':
        print(f'Product of numbers:{x*y}')
    elif  operators=='/':
        print(f'The result of dividing the numbers:{x/y}')
    elif  operators=='%':
        print(f'The remainder of the division of numbers:{x%y}')
    elif  operators=='**':
        print(f'base number:{x**y}')
calculator()
```
