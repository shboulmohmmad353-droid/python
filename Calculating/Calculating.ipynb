### The statistics module in python 
##### استيراد المكتبه 

```python
import statistics
```

### Calculating Mean

```python
data=[1,2,3,4,3,6,7,8,3,2]
mean=statistics.mean(data)
print("Mean:{}".format(mean))
```

```python
mean_data=sum(data)/len(data)
print("Mean:{}".format(mean_data))
```

### Calculating Median

```python
median=statistics.median(data)
print("Median:{}".format(median))
```

### Calculating Mode

```python
mode=statistics.mode(data)
print("Mode:{}".format(mode))
```

### Calculating Variance

```python
variance=statistics.variance(data)
print("Variance:{}".format(variance))
```

### Calculating Stander Deviation

```python
stdev=statistics.stdev(data)
print("Stander Deviation:{}".format(stdev))
```

### Calculating Quartiles

```python
q1=statistics.quantiles(data,n=4)[0]
q3=statistics.quantiles(data,n=4)[-1]
print("First Quartile(Q1):{}".format(q1))
print("Third Quartile(Q1):{}".format(q3))
```

```python
data1=[1,2,3,4,5]
data2=[6,7,8,9,10]
corr_coeff=statistics.correlation(data1,data2)
print("Correlation Cofficient:{}".format(corr_coeff))
```
