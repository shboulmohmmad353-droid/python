### Designing a Neural Networks

```python
import torch.nn as nn
import torch as tr

#create input_tensor with three features
input_tensor=tr.tensor([[0.3471,0.4547,-0.2356]])

#define our linear lyaer
linear_lyaer=nn.Linear(
    in_features=3,

    out_features=2
)

#pass input through linear lyaer
output=linear_lyaer(input_tensor)
print(output)

#weight
print(linear_lyaer.weight)
print("_________________________________________________________________")
#biases
print(linear_lyaer.bias)
```

### First: Binary Classification

```python
# Create input data of shape 5x6
input_data = torch.tensor(
    [[-0.4421, 1.5207, 2.0607, -0.3647, 0.4691, 0.0946],
     [-0.9155, -0.0475, -1.3645, 0.6356, -1.9520, -0.3398],
     [ 0.7406, 1.6763, -0.8511, 0.2432, 0.1123, -0.0653],
     [-1.6630, -0.0718, -0.1285, 0.5396, -0.8288, -0.8622],
     [-0.7413, 1.7920, -0.0883, -0.6685, 0.4745, -0.4245]]
)

#creat binary classification model
model=nn.Sequential(
    nn.Linear(6,4),
    nn.Linear(4,1),
    nn.Sigmoid()

)
print(model)
print("__________________________________________")

#pass input data through model 
output=model(input_data)
print(output)
```
